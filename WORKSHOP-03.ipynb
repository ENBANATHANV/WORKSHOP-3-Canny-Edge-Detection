```python
import cv2
import matplotlib.pyplot as plt

# Load image in grayscale
image = cv2.imread("sam.PNG", 0)

# Check whether image is loaded properly
if image is None:
    print("Error: Image not found")
else:

    # Apply Canny Edge Detection with different parameters
    edges1 = cv2.Canny(image, 50, 100)
    edges2 = cv2.Canny(image, 100, 200)
    edges3 = cv2.Canny(image, 150, 300)

    # Display Original Image
    plt.figure(figsize=(15,10))

    plt.subplot(2,2,1)
    plt.imshow(image, cmap='gray')
    plt.title("Original Image")
    plt.axis('off')

    # Display Edge Detection - Low Threshold
    plt.subplot(2,2,2)
    plt.imshow(edges1, cmap='gray')
    plt.title("Canny Edge Detection (50,100)")
    plt.axis('off')

    # Display Edge Detection - Medium Threshold
    plt.subplot(2,2,3)
    plt.imshow(edges2, cmap='gray')
    plt.title("Canny Edge Detection (100,200)")
    plt.axis('off')

    # Display Edge Detection - High Threshold
    plt.subplot(2,2,4)
    plt.imshow(edges3, cmap='gray')
    plt.title("Canny Edge Detection (150,300)")
    plt.axis('off')

    plt.show()

    # Discussion
    print("\nDiscussion of Results:")
    print("--------------------------------------")
    print("1. Thresholds (50,100):")
    print("   - Detects more edges")
    print("   - Sensitive to noise")
    print("   - Extra unwanted edges may appear\n")

    print("2. Thresholds (100,200):")
    print("   - Balanced edge detection")
    print("   - Detects important edges clearly\n")

    print("3. Thresholds (150,300):")
    print("   - Detects fewer edges")
    print("   - Removes noise effectively")
    print("   - Some weak edges may disappear\n")

    print("Conclusion:")
    print("Canny Edge Detection identifies object boundaries by detecting intensity changes.")
    print("Different threshold values significantly affect edge quality and noise detection.")
```
